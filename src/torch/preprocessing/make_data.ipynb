```python
import re
import numpy as np
import pandas as pd
from collections import defaultdict, deque
```

```python
# df.user_no.unique()
df.sort_values(by=["user_no", "ass_no"], inplace=True) 
```

```python
df.reset_index(inplace=True)
del df['index']
```

```python
df.head()
```

```python
df.head(10)
```

```python
user_dict = defaultdict(list)

for k, v in tqdm(enumerate(df.iterrows())):
    user_dict[v[1]['user_no']].append(v[1]['vec2'])

# print(user_dict[1])
```

### 

### input form

```python
result = defaultdict(list)

for k, v in tqdm(user_dict.items()):
    data = []
    data = deque(data)
    
    for kk, vv in enumerate(v):
        temp = []
        if len(data) > 0:
            # padding 채우기
            if len(data) < 5:
                for _ in range(5 - len(data)):
                    data.appendleft([0 for _ in range(128)])

            temp.append(list(data))
            temp.append(vv)
            
            result[f"{k}_{kk}"].append(temp)
            data.popleft()

        data.append(vv) 
```

```python
df = pd.DataFrame.from_dict(result, orient='index')
```

```python
df.reset_index(inplace=True)
```

```python
def make_user_no(x):
    result = ""
    try:
        result = x[:x.index('_')]
    except Exception as e:
        print(x)
        print(e)
    return result
```

```python
df['user_no'] = df.origin_user_no.apply(make_user_no)
```
