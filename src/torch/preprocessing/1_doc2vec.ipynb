# Preprocessing

```python
import re
import swifter
```

```python
def preprocessing(x):
    result = re.sub(r'\d\)', '', str(x))
    result = re.sub(r'\([^)]*\)', '', str(result))
    result = re.sub('[^0-9a-zA-Zㄱ-힣\s]', '', str(result))
    result = result.replace('\n', ' ').replace('ㅇ ', ' ')
    result = re.sub(' +', ' ', result)
    result = result.strip()
    return result
```

```python
df_origin.drop('Unnamed: 0', axis='columns', inplace=True)
```

```python
# ass_content, ass_detail
df_origin['ass_content'] = df_origin.ass_content.fillna('')
df_origin['ass_detail'] = df_origin.ass_detail.fillna('')
```

```python
df_origin['content5'] = df_origin.b_name + " " + df_origin.agency + " " + df_origin.b_detail + " " + df_origin.ass_name + " " + df_origin.ass_content + " " + df_origin.ass_detail
df_origin['content2'] = df_origin.ass_name + " " + df_origin.agency
```

```python
df_origin.head()
```

```python
df = df_origin.copy()
```

```python
df.head()
```

```python
df['content5'] = df.content5.swifter.apply(preprocessing)
```

```python
df['content2'] = df.content2.swifter.apply(preprocessing)
```

```python
df = df[['year', 'ass_no', 'user_no', 'content5', 'content2']]
```

# doc2vec train

#### data load

```python
import config
from pshmodule.utils import filemanager as fm
```

```python
df = fm.load(config.df_mecab_morphs_path)
```

```python
df.rename(columns={'vec2':'noun2', 'vec5':'noun5'}, inplace=True)
```

```python
df.head()
```

#### train

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
tagged_data2 = [TaggedDocument(words=_d, tags=[str(i)]) for i, _d in enumerate(df.noun2)]
```

```python
tagged_data5 = [TaggedDocument(words=_d, tags=[str(i)]) for i, _d in enumerate(df.noun5)]
```

```python
print(df.noun2[0])
print("--------------------------------------------------------------------------------------------------------------")
print(tagged_data2[0])
```

```python
print(df.noun5[0])
print("--------------------------------------------------------------------------------------------------------------")
print(tagged_data5[0])
```

```python
model5 = Doc2Vec(tagged_data5, vector_size=1024, window=3, min_count=1, workers=4)
# model5 = Doc2Vec(tagged_data5, vector_size=128, window=3, min_count=1, workers=4)
```

```python
model2 = Doc2Vec(tagged_data2, vector_size=1024, window=3, min_count=1, workers=4)
# model2 = Doc2Vec(tagged_data2, vector_size=128, window=3, min_count=1, workers=4)
```

```python
print(model5)
print(model2)
```

#### checkpoint

```python
model5.save(config.doc2vec_morphs+"d2v_5_1024.model")
model2.save(config.doc2vec_morphs+"d2v_2_1024.model")
print("Model Saved")


# model5.save(config.doc2vec_path+"d2v_5.model")
# model2.save(config.doc2vec_path+"d2v_2.model")
# print("Model Saved")
```
