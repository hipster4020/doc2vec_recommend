# data load

```python
import os
import pandas as pd
from tqdm import tqdm
from pshmodule.utils import filemanager as fm
import config
```

```python
file_list = []
data_path = config.data_dir_path
dir_path = data_path
for (root, directories, files) in os.walk(dir_path):
    for file in files:
        file_path = os.path.join(root, file)
        file_list.append(file_path.replace(data_path, ''))
```

```python
file_list
```

```python
for k, v in tqdm(enumerate(file_list)):
    vv = v.replace(".xlsx", "")
    globals()[f"df_{vv}"] = pd.read_excel(data_path + str(v), sheet_name="사업과제정보")
```
